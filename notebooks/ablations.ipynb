```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot style settings
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 11,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

# File paths
paths = {
    "SFF": "/project/pi_wenlongzhao_umass_edu/1/absahoo/experiment_results/prometheusv2/summarize_from_feedback/dataset_size_experiment_results.csv",
    "HelpSteer": "/project/pi_wenlongzhao_umass_edu/1/absahoo/experiment_results/prometheusv2/helpsteer/dataset_size_experiment_results.csv"
}

# Read and aggregate metrics
def extract_metrics(path, metric):
    df = pd.read_csv(path)
    
    grouped = df.groupby(['Model', 'Dataset Size'])[metric].agg(['mean', 'std']).reset_index()
    
    def get_group(model):
        sub = grouped[grouped['Model'] == model].sort_values('Dataset Size')
        return {
            "x": sub['Dataset Size'].tolist(),
            "mean": sub['mean'].tolist(),
            "std": sub['std'].tolist()
        }

    return {
        "ls": get_group('LS'),
        "delta_ls": get_group('Delta LS'),
        "mn": get_group('Multinomial'),
        "delta_mn": get_group('Delta Multinomial')
    }

# Helper to plot mean with std shadow
def plot_with_std(ax, x, mean, std, label, color):
    ax.plot(x, mean, marker='o', label=label, color=color)
    ax.fill_between(x,
                    [m - s for m, s in zip(mean, std)],
                    [m + s for m, s in zip(mean, std)],
                    color=color, alpha=0.2)

# Get metrics
sff_metrics_mse = extract_metrics(paths["SFF"], "MSE")
sff_metrics_acc = extract_metrics(paths["SFF"], "Accuracy")
hs_metrics_mse = extract_metrics(paths["HelpSteer"], "MSE")
hs_metrics_acc = extract_metrics(paths["HelpSteer"], "Accuracy")

# Plot
fig, axs = plt.subplots(1, 4, figsize=(16, 3))
delta_symbol = '\u0394'

# === Plot 1: SFF MSE ===
plot_with_std(axs[0], sff_metrics_mse["ls"]["x"], sff_metrics_mse["ls"]["mean"], sff_metrics_mse["ls"]["std"], 'LS', '#8282ff')
plot_with_std(axs[0], sff_metrics_mse["delta_ls"]["x"], sff_metrics_mse["delta_ls"]["mean"], sff_metrics_mse["delta_ls"]["std"], f"{delta_symbol}-LS", '#fd8585')
axs[0].set_title("Summ. From Feedback")
axs[0].set_xscale('log')
axs[0].set_xlabel("Dataset Size")
axs[0].set_ylabel("Val. MSE")
axs[0].legend()
axs[0].grid(True)

# === Plot 2: SFF Accuracy ===
plot_with_std(axs[1], sff_metrics_acc["mn"]["x"], sff_metrics_acc["mn"]["mean"], sff_metrics_acc["mn"]["std"], 'MN', '#8282ff')
plot_with_std(axs[1], sff_metrics_acc["delta_mn"]["x"], sff_metrics_acc["delta_mn"]["mean"], sff_metrics_acc["delta_mn"]["std"], f"{delta_symbol}-MN", '#fd8585')
axs[1].set_title("Summ. From Feedback")
axs[1].set_xscale('log')
axs[1].set_xlabel("Dataset Size")
axs[1].set_ylabel("Val. Acc.")
axs[1].legend()
axs[1].grid(True)

# === Plot 3: HelpSteer MSE ===
plot_with_std(axs[2], hs_metrics_mse["ls"]["x"], hs_metrics_mse["ls"]["mean"], hs_metrics_mse["ls"]["std"], 'LS', '#8282ff')
plot_with_std(axs[2], hs_metrics_mse["delta_ls"]["x"], hs_metrics_mse["delta_ls"]["mean"], hs_metrics_mse["delta_ls"]["std"], f"{delta_symbol}-LS", '#fd8585')
axs[2].set_title("HelpSteer2")
axs[2].set_xscale('log')
axs[2].set_xlabel("Dataset Size")
axs[2].set_ylabel("Val. MSE")
axs[2].legend()
axs[2].grid(True)

# === Plot 4: HelpSteer Accuracy ===
plot_with_std(axs[3], hs_metrics_acc["mn"]["x"], hs_metrics_acc["mn"]["mean"], hs_metrics_acc["mn"]["std"], 'MN', '#8282ff')
plot_with_std(axs[3], hs_metrics_acc["delta_mn"]["x"], hs_metrics_acc["delta_mn"]["mean"], hs_metrics_acc["delta_mn"]["std"], f"{delta_symbol}-MN", '#fd8585')
axs[3].set_title("HelpSteer2")
axs[3].set_xscale('log')
axs[3].set_xlabel("Dataset Size")
axs[3].set_ylabel("Val. Acc.")
axs[3].legend()
axs[3].grid(True)

# Save and show
plt.tight_layout()
plt.savefig("ratings_dataset_size.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
