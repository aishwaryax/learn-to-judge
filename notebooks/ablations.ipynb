```python
import matplotlib.pyplot as plt
import json

# Plot style settings
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

def load_json(path):
    with open(path, 'r') as f:
        return json.load(f)

# Load your JSONs (dummy paths)
data_json_paths = {
    "Offset_Prometheus": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/reg_data.json",
    "Offset_LLaMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/reg_data.json",
    "Nectar_Prometheus": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/nectar/reg_data.json",
    "Nectar_LLaMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/nectar/reg_data.json"
}

# Load all data
all_data = {key: load_json(path) for key, path in data_json_paths.items()}

# Convert string keys to float and sort
def convert_keys_to_float(d):
    return {float(k): v for k, v in d.items()}

# Assume all datasets share the same x-axis (from BTL keys)
x = sorted([float(k) for k in list(all_data["Offset_Prometheus"]["BTL"].keys())])

def extract_vals(data_dict, method):
    return [data_dict[method][str(k)] for k in x]

# Define styles
delta = '\u0394'
colors = {
    'BTL': color1, f'{delta}-BTL': color2,
    'BTL2': '#77dd77', f'{delta}-BTL2': color3
}
markers = {
    'BTL': 'o', f'{delta}-BTL': 'x',
    'BTL2': 'o', f'{delta}-BTL2': 'x'
}
linestyles = {
    'BTL': '-', f'{delta}-BTL': '-',
    'BTL2': '-', f'{delta}-BTL2': '-'
}

titles = [
    "Offset (Prometheus v2 7B)",
    "Offset (LLaMA 3.1 8B)",
    "Nectar (Prometheus v2 7B)",
    "Nectar (LLaMA 3.1 8B)"
]

# Create 1x4 subplots
fig, axs = plt.subplots(1, 4)

display_name_map = {
    'BTL': 'BTL',
    'Delta BTL': f'{delta}-BTL',
    'BTL2': 'BTL2',
    'Delta BTL2': f'{delta}-BTL2'
}

def plot_all_methods(ax, data, title):
    for method in ['BTL', 'Delta BTL', 'BTL2', 'Delta BTL2']:
        if method not in data:
            continue
        ax.plot(x, extract_vals(data, method),
                label=display_name_map[method],
                color=colors[display_name_map[method]],
                linestyle=linestyles[display_name_map[method]],
                marker=markers[display_name_map[method]])
    ax.set_xscale('log')
    ax.set_xlabel("L2 Strength")
    ax.set_ylabel("Validation Accuracy")
    ax.set_title(title)
    ax.grid(True)

# Plot each dataset
plot_all_methods(axs[0], all_data["Offset_Prometheus"], titles[0])
plot_all_methods(axs[1], all_data["Offset_LLaMA"], titles[1])
plot_all_methods(axs[2], all_data["Nectar_Prometheus"], titles[2])
plot_all_methods(axs[3], all_data["Nectar_LLaMA"], titles[3])

# Layout and legend
plt.tight_layout()
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.10), ncol=4, frameon=False)

# Save and show
plt.savefig("offset_nectar_llama_prometheus_reg_plots.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Plot style settings
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

# File paths
paths = {
    "SFF": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/summarize_from_feedback/dataset_size_experiment_results.csv",
    "HelpSteer": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/helpsteer/dataset_size_experiment_results.csv"
}

paths_llm_scores = {
    "SFF": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/summarize_from_feedback/experiment_results.csv",
    "HelpSteer": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/helpsteer/experiment_results.csv"
}

color_map = {
    "LS": color1,
    "Δ-LS": color2,
    'MN': color4,
    'Δ-MN': color3,
    "Abs LLM Judge": color5
}

def extract_llm_score(path, metric):
    df = pd.read_csv(path)
    row = df[df["Model"] == "LLMRegressor"]
    return row[metric].values[0] if not row.empty else None

llm_scores = {
    "SFF": {
        "MSE": extract_llm_score(paths_llm_scores["SFF"], "MSE"),
        "Accuracy": extract_llm_score(paths_llm_scores["SFF"], "Accuracy")
    },
    "HelpSteer": {
        "MSE": extract_llm_score(paths_llm_scores["HelpSteer"], "MSE"),
        "Accuracy": extract_llm_score(paths_llm_scores["HelpSteer"], "Accuracy")
    }
}

def extract_metrics(path, metric):
    df = pd.read_csv(path)
    df = df[(df['Dataset Size'] >= 10) & (df['Dataset Size'] <= 100)]  # <-- add this line
    grouped = df.groupby(['Model', 'Dataset Size'])[metric].agg(['mean', 'std']).reset_index()
    
    def get_group(model):
        sub = grouped[grouped['Model'] == model].sort_values('Dataset Size')
        return {
            "x": sub['Dataset Size'].tolist(),
            "mean": sub['mean'].tolist(),
            "std": sub['std'].tolist()
        }

    return {
        "ls": get_group('LS'),
        "delta_ls": get_group('Delta LS'),
        "mn": get_group('Multinomial'),
        "delta_mn": get_group('Delta Multinomial')
    }

label_x = "Train Dataset Size"

def plot_with_std(ax, x, mean, std, label, marker):
    ax.plot(x, mean, marker=marker, label=label, color=color_map[label])
    ax.fill_between(x,
                    [m - s for m, s in zip(mean, std)],
                    [m + s for m, s in zip(mean, std)],
                    color=color_map[label], alpha=0.2)

# Get metrics
sff_metrics_mse = extract_metrics(paths["SFF"], "MSE")
sff_metrics_acc = extract_metrics(paths["SFF"], "Accuracy")
hs_metrics_mse = extract_metrics(paths["HelpSteer"], "MSE")
hs_metrics_acc = extract_metrics(paths["HelpSteer"], "Accuracy")

fig, axs = plt.subplots(1, 4, figsize=(16, 3))
delta_symbol = '\u0394'

# === Plot 1: SFF MSE ===
plot_with_std(axs[0], sff_metrics_mse["ls"]["x"], sff_metrics_mse["ls"]["mean"], sff_metrics_mse["ls"]["std"], 'LS', 'o')
plot_with_std(axs[0], sff_metrics_mse["delta_ls"]["x"], sff_metrics_mse["delta_ls"]["mean"], sff_metrics_mse["delta_ls"]["std"], 'Δ-LS', 'x')
axs[0].set_title("Summ. From Feedback")
axs[0].axhline(llm_scores["SFF"]["MSE"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[0].set_xlabel(label_x)
axs[0].set_ylabel("Val. MSE")
axs[0].grid(True)

# === Plot 2: SFF Accuracy ===
plot_with_std(axs[1], sff_metrics_acc["mn"]["x"], sff_metrics_acc["mn"]["mean"], sff_metrics_acc["mn"]["std"], 'MN', 'o')
plot_with_std(axs[1], sff_metrics_acc["delta_mn"]["x"], sff_metrics_acc["delta_mn"]["mean"], sff_metrics_acc["delta_mn"]["std"], 'Δ-MN', 'x')
axs[1].set_title("Summ. From Feedback")
axs[1].axhline(llm_scores["SFF"]["Accuracy"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[1].set_xlabel(label_x)
axs[1].set_ylabel("Val. Acc.")
axs[1].grid(True)

# === Plot 3: HelpSteer MSE ===
plot_with_std(axs[2], hs_metrics_mse["ls"]["x"], hs_metrics_mse["ls"]["mean"], hs_metrics_mse["ls"]["std"], 'LS', 'o')
plot_with_std(axs[2], hs_metrics_mse["delta_ls"]["x"], hs_metrics_mse["delta_ls"]["mean"], hs_metrics_mse["delta_ls"]["std"], 'Δ-LS', 'x')
axs[2].set_title("HelpSteer2")
axs[2].axhline(llm_scores["HelpSteer"]["MSE"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[2].set_xlabel(label_x)
axs[2].set_ylabel("Val. MSE")
axs[2].grid(True)

# === Plot 4: HelpSteer Accuracy ===
plot_with_std(axs[3], hs_metrics_acc["mn"]["x"], hs_metrics_acc["mn"]["mean"], hs_metrics_acc["mn"]["std"], 'MN', 'o')
plot_with_std(axs[3], hs_metrics_acc["delta_mn"]["x"], hs_metrics_acc["delta_mn"]["mean"], hs_metrics_acc["delta_mn"]["std"], 'Δ-MN', 'x')
axs[3].set_title("HelpSteer2")
axs[3].axhline(llm_scores["HelpSteer"]["Accuracy"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[3].set_ylabel("Val. Acc.")
axs[3].grid(True)

ticks = np.arange(0, 101, 20)
for ax in axs:
    ax.set_xticks(ticks)

# Save and show
plt.tight_layout()
handles1, labels1 = axs[0].get_legend_handles_labels()
handles2, labels2 = axs[1].get_legend_handles_labels()
handles, labels = handles1[:-1] + handles2, labels1[:-1] + labels2
fig.legend(
    handles, labels,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.05),
    ncol=len(handles),
    frameon=False,
    borderaxespad=0.0
)
plt.savefig("ratings_dataset_size_prometheus.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Plot style settings
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

# File paths
paths = {
    "SFF": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/llama3.18b/summarize_from_feedback/dataset_size_experiment_results.csv",
    "HelpSteer": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/llama3.18b/helpsteer/dataset_size_experiment_results.csv"
}

paths_llm_scores = {
    "SFF": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/llama3.18b/summarize_from_feedback/experiment_results.csv",
    "HelpSteer": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/llama3.18b/helpsteer/experiment_results.csv"
}

color_map = {
    "LS": color1,
    "Δ-LS": color2,
    'MN': color4,
    'Δ-MN': color3,
    "Abs LLM Judge": color5
}

def extract_llm_score(path, metric):
    df = pd.read_csv(path)
    row = df[df["Model"] == "LLMRegressor"]
    return row[metric].values[0] if not row.empty else None

llm_scores = {
    "SFF": {
        "MSE": extract_llm_score(paths_llm_scores["SFF"], "MSE"),
        "Accuracy": extract_llm_score(paths_llm_scores["SFF"], "Accuracy")
    },
    "HelpSteer": {
        "MSE": extract_llm_score(paths_llm_scores["HelpSteer"], "MSE"),
        "Accuracy": extract_llm_score(paths_llm_scores["HelpSteer"], "Accuracy")
    }
}

def extract_metrics(path, metric):
    df = pd.read_csv(path)
    df = df[(df['Dataset Size'] >= 10) & (df['Dataset Size'] <= 100)]
    grouped = df.groupby(['Model', 'Dataset Size'])[metric].agg(['mean', 'std']).reset_index()
    
    def get_group(model):
        sub = grouped[grouped['Model'] == model].sort_values('Dataset Size')
        return {
            "x": sub['Dataset Size'].tolist(),
            "mean": sub['mean'].tolist(),
            "std": sub['std'].tolist()
        }

    return {
        "ls": get_group('LS'),
        "delta_ls": get_group('Delta LS'),
        "mn": get_group('Multinomial'),
        "delta_mn": get_group('Delta Multinomial')
    }

label_x = "Train Dataset Size"

def plot_with_std(ax, x, mean, std, label, marker):
    ax.plot(x, mean, marker=marker, label=label, color=color_map[label])
    ax.fill_between(x,
                    [m - s for m, s in zip(mean, std)],
                    [m + s for m, s in zip(mean, std)],
                    color=color_map[label], alpha=0.2)

# Get metrics
sff_metrics_mse = extract_metrics(paths["SFF"], "MSE")
sff_metrics_acc = extract_metrics(paths["SFF"], "Accuracy")
hs_metrics_mse = extract_metrics(paths["HelpSteer"], "MSE")
hs_metrics_acc = extract_metrics(paths["HelpSteer"], "Accuracy")

fig, axs = plt.subplots(1, 4, figsize=(16, 3))
delta_symbol = '\u0394'

# === Plot 1: SFF MSE ===
plot_with_std(axs[0], sff_metrics_mse["ls"]["x"], sff_metrics_mse["ls"]["mean"], sff_metrics_mse["ls"]["std"], 'LS', 'o')
plot_with_std(axs[0], sff_metrics_mse["delta_ls"]["x"], sff_metrics_mse["delta_ls"]["mean"], sff_metrics_mse["delta_ls"]["std"], 'Δ-LS', 'x')
axs[0].set_title("Summ. From Feedback")
axs[0].axhline(llm_scores["SFF"]["MSE"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[0].set_xlabel(label_x)
axs[0].set_ylabel("Val. MSE")
axs[0].grid(True)

# === Plot 2: SFF Accuracy ===
plot_with_std(axs[1], sff_metrics_acc["mn"]["x"], sff_metrics_acc["mn"]["mean"], sff_metrics_acc["mn"]["std"], 'MN', 'o')
plot_with_std(axs[1], sff_metrics_acc["delta_mn"]["x"], sff_metrics_acc["delta_mn"]["mean"], sff_metrics_acc["delta_mn"]["std"], 'Δ-MN', 'x')
axs[1].set_title("Summ. From Feedback")
axs[1].axhline(llm_scores["SFF"]["Accuracy"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[1].set_xlabel(label_x)
axs[1].set_ylabel("Val. Acc.")
axs[1].grid(True)

# === Plot 3: HelpSteer MSE ===
plot_with_std(axs[2], hs_metrics_mse["ls"]["x"], hs_metrics_mse["ls"]["mean"], hs_metrics_mse["ls"]["std"], 'LS', 'o')
plot_with_std(axs[2], hs_metrics_mse["delta_ls"]["x"], hs_metrics_mse["delta_ls"]["mean"], hs_metrics_mse["delta_ls"]["std"], 'Δ-LS', 'x')
axs[2].set_title("HelpSteer2")
axs[2].axhline(llm_scores["HelpSteer"]["MSE"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[2].set_xlabel(label_x)
axs[2].set_ylabel("Val. MSE")
axs[2].grid(True)

# === Plot 4: HelpSteer Accuracy ===
plot_with_std(axs[3], hs_metrics_acc["mn"]["x"], hs_metrics_acc["mn"]["mean"], hs_metrics_acc["mn"]["std"], 'MN', 'o')
plot_with_std(axs[3], hs_metrics_acc["delta_mn"]["x"], hs_metrics_acc["delta_mn"]["mean"], hs_metrics_acc["delta_mn"]["std"], 'Δ-MN', 'x')
axs[3].set_title("HelpSteer2")
axs[3].axhline(llm_scores["HelpSteer"]["Accuracy"], linestyle='--', color=color_map["Abs LLM Judge"], label='Abs LLM Judge')
axs[3].set_xlabel(label_x)
axs[3].set_ylabel("Val. Acc.")
axs[3].grid(True)

ticks = np.arange(0, 101, 20)
for ax in axs:
    ax.set_xticks(ticks)
# Save and show
plt.tight_layout()
handles1, labels1 = axs[0].get_legend_handles_labels()
handles2, labels2 = axs[1].get_legend_handles_labels()
handles, labels = handles1[:-1] + handles2, labels1[:-1] + labels2
fig.legend(
    handles, labels,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.05),
    ncol=len(handles),
    frameon=False,
    borderaxespad=0.0
)
plt.savefig("ratings_dataset_size_llama.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Plot style settings
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

delta = '\u0394'

# File paths
paths = {
    "OFFSET_LLAMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/dataset_size_experiment_results.csv",
    "OFFSET_PROMETHEUS": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/dataset_size_experiment_results.csv",
    "NECTAR_LLAMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/dataset_size_experiment_results.csv",
    "NECTAR_PROMETHEUS": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/dataset_size_experiment_results.csv",
}

paths_llm_scores = {
    "OFFSET_LLAMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/experiment_results.csv",
    "OFFSET_PROMETHEUS": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/experiment_results.csv",
    "NECTAR_LLAMA": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/experiment_results.csv",
    "NECTAR_PROMETHEUS": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/offset_bias/experiment_results.csv"
}

def extract_llm_score(path, metric, key):
    df = pd.read_csv(path)
    row = df[df["Model"] == key]
    return row[metric].values[0] if not row.empty else None

llm_scores = {
    key: {
        "Rel LLM Judge": extract_llm_score(paths_llm_scores[key], "Accuracy", "LLMRegressor"),
        "Abs LLM Judge": extract_llm_score(paths_llm_scores[key], "Accuracy", "LLM2Regressor")
    }
    for key in paths_llm_scores
}

def plot_llm_references(ax, dataset_key):
    colors = {
        "Abs LLM Judge": color5,  
        "Rel LLM Judge": color6,
    }
    for name, color in colors.items():
        value = llm_scores[dataset_key][name]
        if value is not None:
            ax.axhline(value, linestyle='--', color=color, label=name)

# Extract and aggregate metrics
def extract_metrics(path, metric):
    df = pd.read_csv(path)
    df = df[(df['Dataset Size'] >= 10) & (df['Dataset Size'] <= 100)]
    grouped = df.groupby(['Model', 'Dataset Size'])[metric].agg(['mean', 'std']).reset_index()

    def get_group(model_name):
        sub = grouped[grouped['Model'] == model_name].sort_values('Dataset Size')
        return {
            "x": sub['Dataset Size'].tolist(),
            "mean": sub['mean'].tolist(),
            "std": sub['std'].tolist()
        }

    return {
        "BTL": get_group('BTL'),
        "BTL2": get_group('BTL2'),
        f'{delta}-BTL': get_group(f'Delta BTL'),
        f'{delta}-BTL2': get_group(f'Delta BTL2'),
    }

label_x = "Train Dataset Size"
# Plot helper
def plot_all_methods(ax, metrics, title, dataset_key):
    colors = {
        'BTL': color1,              # pastel blue
        f'{delta}-BTL': color2,        # pastel pink
        'BTL2': color4,             # pastel green
        f'{delta}-BTL2': color3        # red (unchanged)
    }
    markers = {
        'BTL': 'o',             
        f'{delta}-BTL': 'x',       
        'BTL2': 'o',             
        f'{delta}-BTL2': 'x'        
    }
    for method, data in metrics.items():
        ax.plot(data["x"], data["mean"], marker=markers[method], label=method, color=colors[method])
        ax.fill_between(data["x"],
                        [m - s for m, s in zip(data["mean"], data["std"])],
                        [m + s for m, s in zip(data["mean"], data["std"])],
                        alpha=0.2, color=colors[method])
    plot_llm_references(ax, dataset_key)
    ax.set_title(title)
    ax.set_xlabel(label_x)
    ticks = np.arange(0, 101, 20)
    ax.set_xticks(ticks)
    ax.set_ylabel("Val. Acc.")
    ax.grid(True)

# Read data
offset_llama_metrics = extract_metrics(paths["OFFSET_LLAMA"], "Accuracy")
offset_prometheus_metrics = extract_metrics(paths["OFFSET_PROMETHEUS"], "Accuracy")
nectar_llama_metrics = extract_metrics(paths["NECTAR_LLAMA"], "Accuracy")
nectar_prometheus_metrics = extract_metrics(paths["NECTAR_PROMETHEUS"], "Accuracy")

# Plot
fig, axs = plt.subplots(1, 4, figsize=(16, 3))

plot_all_methods(axs[0], offset_prometheus_metrics, "Offset Bias | Prometheus", "OFFSET_PROMETHEUS")
plot_all_methods(axs[1], offset_llama_metrics, "Offset Bias | LLaMA", "OFFSET_LLAMA")
plot_all_methods(axs[2], nectar_prometheus_metrics, "Nectar | Prometheus", "NECTAR_PROMETHEUS")
plot_all_methods(axs[3], nectar_llama_metrics, "Nectar | LLaMA", "NECTAR_LLAMA")

plt.tight_layout()
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(
    handles, labels,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.05),
    ncol=len(handles),
    frameon=False,
    borderaxespad=0.0
)
plt.savefig("preferences_dataset_size_llama_prometheus.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "figure.figsize": (16, 3),
    "pdf.fonttype": 42
})

# Constants
colors = [color1, color3]  # Prometheus2 7B, MiniLM
metrics = ['MSE', 'Accuracy']
delta_names = {
    "LS": "LS",
    "Multinomial": "MN",
    "Delta LS": "Δ-LS",
    "Delta Multinomial": "Δ-MN",
}

# File paths and task names
tasks = [
    ("Summ. from Feedback", {
        "prometheus": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/summarize_from_feedback/pca_experiment_results.csv",
        "minilm": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/summarize_from_feedback/sentence_embeddings_experiment_results.csv"
    }),
    ("HelpSteer", {
        "prometheus": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/helpsteer/pca_experiment_results.csv",
        "minilm": "/project/pi_wenlongzhao_umass_edu/1/experiment_results/prometheusv2/helpsteer/sentence_embeddings_experiment_results.csv"
    }),
]

# Set up 1x4 subplots
fig, axs = plt.subplots(1, 4, figsize=(16, 3))
width = 0.35
plot_idx = 0

for task_name, paths in tasks:
    df_prometheus = pd.read_csv(paths["prometheus"]).iloc[:4].copy()
    df_minilm = pd.read_csv(paths["minilm"]).iloc[:4].copy()

    df_prometheus["Model"] = df_prometheus["Model"].apply(lambda x: delta_names.get(x, x))
    df_minilm["Model"] = df_minilm["Model"].apply(lambda x: delta_names.get(x, x))
    models = df_prometheus["Model"].values
    x = np.arange(len(models))

    for metric in metrics:
        ax = axs[plot_idx]
        ax.bar(x - width/2, df_prometheus[metric], width, label='Prometheus2-7B', color=colors[0])
        ax.bar(x + width/2, df_minilm[metric], width, label='MiniLM', color=colors[1])

        ax.set_title(f"{metric} ({task_name})")
        ax.set_xticks(x)
        ax.set_xticklabels(models, rotation=0)
        if plot_idx == 0:
            ax.set_ylabel("Metric Value")
        plot_idx += 1

plt.tight_layout()
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(
    handles, labels,
    loc='upper center',
    bbox_to_anchor=(0.5, 1.05),
    ncol=len(handles),
    frameon=False,
    borderaxespad=0.0
)
plt.savefig("sentence_embedding_ablation_mse_acc_ratings.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
